```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### basic graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your graph(dpi specifies pixels per inch. when saving probabbly should use 300 if possible)
plt.figure(figsize=(5,4), dpi=300)

# keyword argument notation
# plt.plot(x,y, label='2x', color='red', linewidth=2 , marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

#use shortern notation
#frmt = '[color][marker][line]'
plt.plot(x,y, 'r.--' ,label='2x')

##Line 2

# select intervals we want to plot points at
x2 = np.arange(0,4.5,0.5)

# plot part of the graph as line 
plt.plot(x2[:6], x2[:6]**2, label='x^2')

# plot reminder of graph as a dot
plt.plot(x2[5:], x2[5:]**2,'--')

# add a title(specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname':'Comic Sans MS', 'fontsize':15})

# x and y axis lables
plt.xlabel('X axis')
plt.ylabel('Y axis')

# x,y axis tickmarkers(scale of your graph)
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8])

# add a legend
plt.legend()

# save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# show plot
plt.show()
```
