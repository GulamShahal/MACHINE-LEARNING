```python
df_country = pd.read_excel('country-code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country, on='Country Code',how='left')
final_df.head(2)
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_val= final_df.Country.value_counts().values
```

```python
#pie chart - Top 3 countries that use zomato
plt.pie(country_val[:3],labels=country_name[:3],autopct = '%1.2f%%')
```

observation : Zomato maximum transaction are from india after that USA and then untited kingdoms

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## observation
1. when rating is between 4.5 to 4.9 - excellent
2. when rating is between 4.0 to 4.5 - very good
3. when rating is between 3.5 to 3.9 - good
4. when rating is between 3.0 to 3.4 - average
5. when rating is between 2.5 to 2.9 - average
6. when rating is between 2.0 to 2.4 - poor

```python
import matplotlib as mt
mt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x = 'Aggregate rating', y = 'Rating count', data = ratings)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating count', hue = 'Rating color',data = ratings, palette = ['white','red','orange','yellow','green','green'])
```

## observation
1. not rated count is very high
2. maximium number of raiting are in beteween 2.5 to 3.5

```python
## count plot
sns.countplot(x='Rating color', data=ratings, palette =  ['white','red','orange','yellow','green','green'])
```
