```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### basic graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label='2x', color='red')

plt.title('Our First Graph!', fontdict={'fontname':'Comic Sans MS', 'fontsize':15})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.show()
```
